```python
import numpy as np 
import pandas as pd
```

```python
class My_Linear_Regression:
    def __init__(self,learning_rate=0.01,epochs=1000):
        self.learning_rate = learning_rate 
        self.epochs = epochs 
        self.w = 0 
        self.b = 0 

    def fit(self,X,y):
        self.w = np.zeros(X.shape[1])
        self.b = 0 
        for _ in range(self.epochs):
            y_pred = np.dot(X,self.w)+self.b
            error = y_pred - y 
            dw = (2/X.shape[0])*np.dot(X.T,error)
            db = (2/X.shape[0])*np.sum(error)
            self.w -= self.learning_rate*dw 
            self.b -= self.learning_rate*db 
        return self 

    def predict(self,X):
        return  np.dot(X,self.w)+self.b  

```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import numpy as np

# Dummy Dataset
X = np.array([
    [1],
    [2],
    [3],
    [4],
    [5]
])

y = np.array([2,4,6,8,10])

# My Model
my_model = My_Linear_Regression(
    learning_rate=0.01,
    epochs=1000
)

my_model.fit(X,y)

my_pred = my_model.predict(X)

print(my_model.w)
print(my_model.b)
print(my_pred)

# Sklearn
sk_model = LinearRegression()

sk_model.fit(X,y)

sk_pred = sk_model.predict(X)

print(sk_model.coef_)
print(sk_model.intercept_)
print(sk_pred)

print(r2_score(y,my_pred))
print(r2_score(y,sk_pred))
```
